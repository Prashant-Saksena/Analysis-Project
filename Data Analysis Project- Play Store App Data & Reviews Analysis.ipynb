# Treating the null values

```python
# - We have two options- 

# - 1- Drop the null values  - If any column has > 25% Null values, then we DROP that column
#                            - Nature of the column (Numeric continuous / Categorical)

# - Drop that particular row - if its not highly affecting you data.

# - 2- Impute the null values - We always impute the null values columnwise
```

```python
# Rating Col.

df['Rating']   # --> is SAME as  "df.Rating"  --> This methon works ONLY when there is "single word", if there are 
#                                                 multiple words, we have to use standard [] method - df['Rating']
#     OR   we can replace the spaces in col header names with an underscore    df.columns=df.columns.str.replace(' ','_')
```

```python
df['Rating'].isnull().sum()
```

```python
# - OR - 

df.Rating.isna().sum()
```

```python
df['Rating'].count()
```

```python
# %age of null values in Rating Col

Rating_Null_percentage = df['Rating'].isnull().sum() *100/df['Rating'].count()

print(Rating_Null_percentage)
```

```python
# RETAIN the Rating Column, CAN NOT Drop it, since Null values %age is < 25%.

# We will impute the null values in Rating Col
```

```python
df.Rating.unique()
```

```python
# - It's a cataegorical column
# - We can fill the ull values with the MODE value
```

```python
df.Rating.value_counts()
```

```python
df.Rating.mode()
```

```python
df.Rating.mode()[0]
```

```python
# Filling null values in Rating Col with the MODE value of that column

df.Rating.fillna(df.Rating.mode()[0], inplace= True)

df.isnull().sum()    # - Validating if null values are removed from Rating Column.
```

```python
# Defining a function to replace all null values in a column with the MODE value of that column

def fill_na(col):
    return df[col].fillna(df[col].mode()[0], inplace = True)

for i in df.columns:
    fill_na(i)
```

```python
df.isnull().sum()      # --> No null values r present in any column now.
```

### Reviews - Outlier Detection

```python
q1 = df.Reviews.quantile(0.25)
q3 = df.Reviews.quantile(0.75)

print(f'q1 = {q1}')
print(f'q3 = {q3}')
```

```python
# IQR Method

IQR = q3-q1
print(f'IQR = {IQR}')
```

```python
Upper_limit = q3 + 1.5*IQR
Lower_limit = q1 - 1.5*IQR

print(f'Upper Limit is: {Upper_limit}')

print(f'Lower Limit is: {Lower_limit}')
```

```python
# Data in Outlier Range

df.loc[df.Reviews > Upper_limit]
```
